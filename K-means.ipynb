```r
summary(DATA)
```

```r
DATA_VAR <- c("sex","age","emplnw","emptype1","inc")
x <- DATA[DATA_VAR]
x
```

```r
s_data<-na.omit(x)
nrow(s_data)
s_data
```

```r

x1<-s_data[s_data$inc!=98 & s_data$inc!=99,] 
y<-x1[x1$age!=98 & x1$age!=99,] 
z<-y[y$emplnw!=98 & y$emplnw!=99,] 
cleaned_data<-z[z$emptype1!=98 & z$emptype1!=99,]
nrow(cleaned_data)
colnames(cleaned_data)[3]<-"Employment_Type" 
colnames(cleaned_data)[4]<-"Employment_Place"
colnames(cleaned_data)[5]<-"Income"
cleaned_data
```

```r
summary(cleaned_data)
```

```r
set.seed(1000)
kmeans3<-kmeans(cleaned_data,centers=3,algorithm=c("Hartigan-Wong"))
kmeans3
```

```r
kmeans3$size
```

```r
set.seed(1000)
kmeans6<-kmeans(cleaned_data,centers=6,algorithm=c("Hartigan-Wong"))
kmeans6
```
