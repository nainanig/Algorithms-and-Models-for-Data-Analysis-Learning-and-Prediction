```r
library("class")
```

```r
data_read=read.csv('/home/naina/Downloads/german_credit_data1.csv')
```

```r
data_read
```

```r
num=sapply(data_read,as.numeric)
num
```

```r
x1 <- replace(num,is.na(num),0)
x1
```

```r
n = function(x){return((x-min(x))/(max(x)-min(x)))}    #referred http://vitalflux.com/data-science-scale-normalize-numeric-data-using-r/
```

```r
df=data.frame(x1)
dfNorm <- as.data.frame(lapply(df, n))
dfNorm
```

```r
dfNorm=as.data.frame(dfNorm)
```

```r
set.seed(126)
n.points=1000
sampling.rate=0.75
set_labels=n.points*(1-sampling.rate)
training<-sample(1:n.points, sampling.rate*n.points, replace=FALSE)
train<- subset(dfNorm[training,],select=c(Age,Sex,Job,Housing,Saving.accounts,Checking.account,Credit.amount,Duration,Purpose))

testing<-setdiff(1:n.points,training)
test<-subset(dfNorm[testing,], select=c(Age,Sex,Job,Housing,Saving.accounts,Checking.account,Credit.amount,Duration,Purpose))

cl<- dfNorm$Credit.Risks[training]

true.labels<-dfNorm$Credit.Risks[testing]
```

```r
knn(train,test,cl,k=3) 
```

```r
for(i in 1:20) {
  print(i)
  predicted.labels <- knn(train,test,cl,i)
  num.incorrect.labels<-sum(predicted.labels !=true.labels)
  misclassification.rate_1<-num.incorrect.labels/set_labels
  print(misclassification.rate_1)
    
}
```
